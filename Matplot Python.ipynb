```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')

#seaborn is a default loaded data set 
```

```python
tips.head()
```

```python
#what is distribution 
# spread of the value is distrubution. so hence the distribution is and should and has to be univariant

#sns.displot(tips['total_bill'])
sns.displot(tips['total_bill'],bins=30)
# Y axis is the count, number of people vs Y axis has teh total bill value
# if we increase the Bins to maore values then the distribution increases

```

```python
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x='total_bill',y='tip',data =tips)
#this is used for showing the relationbetween the 2 variables. 
```

```python
sns.jointplot(x='total_bill',y='tip',data =tips,kind='hex')
#jointplot uses to relate between the numerical colusmns
#hex kind gives the view of the more denser data in the graph. If we see from the graph that the data is more denser between 10 to 20
```

```python
tips.info()
```

```python
sns.pairplot(tips)
#takes on the Numerical columns only
```

```python
sns.rugplot(tips['total_bill'])
#This gives the density indicator and if we take each points , we can draw thw uniform
# distrubution with it. Jus like showin in 123., ie,, from 1st point near 0.06 to second one 
# with the distruibution curve
```

```python
sns.displot(tips['total_bill'],kde=True,bins=30)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#this is mostly like X axis will be categorical and the Y axis is Numerical 
#This is plotted mostly on the mean values of the numerical data
#By default it plots till the mean value 
#the max value of the total bill is actually 50.10
```

```python
sns.countplot(x='sex',data=tips)
# this is for the categorical data and is univariant 
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#box plot is used for identifying the outliers 
# We could see that foor all the days there is outliers present 
#this also gives the distribution and as well as the mean value as well 
# thurrsday the more is on the 10 to 20 and the mean is around 15 
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
#gives more details on the gender when you give hue 
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='coolwarm')
#to give color 
```

```python
sns.violinplot(x='day',y='total_bill',data = tips)
#By the fatnesss we can see where the data is more denser
```

```python
sns.violinplot(x='day',y='total_bill',data = tips,hue='sex',palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data = tips,hue='sex',palette='coolwarm',split=True)
#comes in the same violin for both male and female
```
